```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Carregar udio (exemplo: substitua pelo seu arquivo .wav ou .mp3)
y, sr = librosa.load('Verrina - 7 anos mix2026 V1.wav', sr=None)

# Data Understanding: Visualizar waveform e espectrograma
plt.figure(figsize=(14, 5))
librosa.display.waveshow(y, sr=sr)
plt.title('Waveform - Anlise Exploratria')
plt.show()

# Espectrograma Mel (como anlise de frequncias)
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
S_dB = librosa.power_to_db(S, ref=np.max)
plt.figure(figsize=(14, 5))
librosa.display.specshow(S_dB, x_axis='time', y_axis='mel', sr=sr)
plt.colorbar(format='%+2.0f dB')
plt.title('Espectrograma Mel - Identificando mscaras frequenciais')
plt.show()
```
